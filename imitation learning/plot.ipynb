```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Create x values
x = np.linspace(-6, 6, 1000)

# Parameters for distributions
mu1 = 0  # mean for single Gaussian
sigma1 = 1.0  # std for single Gaussian

mu2_1, mu2_2 = -2, 0.1  # means for bimodal
sigma2 = 0.3  # low std for bimodal
w1, w2 = 0.5, 0.5  # weights for bimodal

# Calculate distributions
gaussian = (1/(sigma1 * np.sqrt(2*np.pi))) * np.exp(-(x-mu1)**2/(2*sigma1**2))
bimodal = w1 * (1/(sigma2 * np.sqrt(2*np.pi))) * np.exp(-(x-mu2_1)**2/(2*sigma2**2)) + \
          w2 * (1/(sigma2 * np.sqrt(2*np.pi))) * np.exp(-(x-mu2_2)**2/(2*sigma2**2))

# Create plot
plt.figure(figsize=(10, 6))
plt.plot(x, gaussian, 'b-', label='Actual', alpha=0.7, linewidth=2)
plt.plot(x, bimodal, 'r-', label='Sampled', alpha=0.7, linewidth=2)

#plt.title('Gaussian vs Bimodal Gaussian Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid(True, alpha=0.3)
plt.legend()

# Set y-axis limit to focus on the important parts
plt.ylim(0, max(np.max(gaussian), np.max(bimodal)) * 1.1)

plt.show()
```
